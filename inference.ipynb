## Inferencing Mean

bagaimana mengambil kesimpulan nilai dari Mean of Population

kita mau tau, berapa sih rata2 populasi?

```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('lady_tast.csv')
```

```python
df.numbers.plot(kind = 'hist')
plt.show()
```

```python
df.numbers.mean()
```

## CLT

```python
all_means = []
for i in range(1000):
    ## Ambil dan hitung rata2-nya
    mu = df.sample(10)['numbers'].mean()
    
    ## simpan rata2-nya
    all_means.append(mu)
```

```python
## kemungkinan populasi ber rata-rata diatas 80 adalah kecil
## kemungkinan populasi ber rata-rata dibawah 40 adalah kecil
```

```python
sns.distplot(all_means)
plt.show()
```

```python
mu_mean = pd.Series(all_means).mean()
std_mean = pd.Series(all_means).std()

all_mean_standar = (np.array(all_means)-mu_mean)/std_mean
```

```python
sns.distplot(all_mean_standar)
plt.show()
```
